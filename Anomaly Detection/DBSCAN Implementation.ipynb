```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
%matplotlib inline
```

```python
from sklearn.datasets import make_circles
from sklearn.cluster import DBSCAN
```

```python
X,y = make_circles(n_samples = 750, noise = 0.1,factor=0.3)
```

```python
X
```

```python
plt.scatter(X[:,0], X[:,1],c=y)
```

```python
DBSCAN_model = DBSCAN(eps=0.10)
```

```python
DBSCAN_model.fit_predict(X)
```

```python
DBSCAN_model.labels_
```

```python
plt.scatter(X[:,0], X[:,1], c=DBSCAN_model.labels_, cmap='viridis')
```
