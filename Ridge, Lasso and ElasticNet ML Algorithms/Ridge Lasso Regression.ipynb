```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset = pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv',header = 1)
```

```python
dataset 
```

```python
dataset.info()
```

## Data Cleaning

```python
## Missing values

dataset[dataset.isnull().any(axis=1)]
```

The Dataset is converted into two set based on Region from index 122, we can make a new column based on the region
1. Begaia Region Dataset
2. Sidi-Bel Abbes Region Dataset

So Add a new Column with region

```python
dataset.loc[:122,"Region"] = 0
dataset.loc[122:,"Region"] = 1
```

```python
df = dataset
df.info()
```

```python
df[['Region']] = df[['Region']].astype('int')
```

```python
df
```

```python
df.isnull().sum()
```

```python
## drop the columns with null values

df = df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]] # this row is not needed anymore
```

```python
df = df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns # there are unwanted spaces in the column names which need to be removed
```

```python
## fixing spaces in column names

df.columns = df.columns.str.strip() # remove leading and trailing spaces
df.columns
```

```python
df.info() 
```

#### changing to data type of columns based on there format

```python
df[['day','month','year','Temperature','RH','Ws']] = df[['day','month','year','Temperature','RH','Ws']].astype('int')
```

```python
df[['Rain','FFMC','DMC','DC','ISI','BUI','FWI']] = df[['Rain','FFMC','DMC','DC','ISI','BUI','FWI']].astype('float')
```

```python
df.info()
```

```python
df.describe()
```

```python
# Let's save the cleaned dataset to a new CSV file
df.to_csv('Algerian_forest_fires_dataset_cleaned.csv', index=False)
```

## Exploratory Data Analysis

```python
# drop the day month year columns
df_copy = df.drop(['day','month','year'],axis = 1)
```

```python
df_copy
```

```python
# encoding of the categorical variables i.e., classes

df_copy['Classes'] = np.where(df_copy['Classes'].str.contains('not fire'), 0, 1)
```

```python
df_copy.head()
```

```python
df_copy.tail()
```

```python
df_copy['Classes'].value_counts()
```

```python
# plot density of all the features

plt.style.use('seaborn-v0_8')
df_copy.hist(bins=50, figsize=(20, 15))
plt.show()
```

```python
# Percentage for pie chart

percentage = df_copy['Classes'].value_counts(normalize=True) * 100
```

```python
# plotting pie chart
classlabels = ['Fire', 'Not Fire']
plt.figure(figsize=(12,7))
plt.pie(percentage,labels = classlabels,autopct='%1.1f%%')
plt.title('Pie Chart of Classes')   
plt.show()
```

```python
## Correlation matrix

df_copy.corr()
```

```python
# heatmap

sns.heatmap(df_copy.corr())
plt.show()
```

```python
## boxplot

sns.boxplot(df['FWI'])
plt.title('Boxplot of FWI')
plt.show()
```

```python
df['Classes'] = np.where(df['Classes'].str.contains('not fire'), 'not fire', 'fire')
```

```python
# Monthly Fire Analysis

dftemp = df.loc[df['Region'] == 1]
plt.subplots(figsize = (13,6))
sns.set_style('whitegrid')
sns.countplot(x='month',hue = 'Classes',data= df)
plt.title('Fire Analysis of Sidi Bel Abbes Region')
plt.xlabel('Month',weight = 'bold')
plt.ylabel('Number of Fires',weight = 'bold')
plt.show()
```

```python
# Monthly Fire Analysis

dftemp = df.loc[df['Region'] == 10]
plt.subplots(figsize = (13,6))
sns.set_style('whitegrid')
sns.countplot(x='month',hue = 'Classes',data= df)
plt.title('Fire Analysis of Brjaia Region')
plt.xlabel('Month',weight = 'bold')
plt.ylabel('Number of Fires',weight = 'bold')
plt.show()
```

It is observed that August and September had the most number of forest fires for both region. And from above plots we can understand few things. 

Most of the fire happened in August and very high fires happened in only three months - June, July and August

Less fire was on September
