```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## let create a synthetic dataset
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples=1000,n_features = 2, n_classes=2, 
                          n_clusters_per_class=2,n_redundant=0)
```

```python
X
```

```python
y
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0],y=pd.DataFrame(X)[1],hue=y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=10)
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(kernel='linear') # if our data is clearly separable then we can use linear kernel
```

```python
svc.fit(X_train, y_train)
```

```python
y_pred = svc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
rbf = SVC(kernel='rbf')
```

```python
rbf.fit(X_train, y_train)
```

```python
y_pred_rbf = rbf.predict(X_test)
print("RBF Kernel Accuracy:", accuracy_score(y_test, y_pred_rbf))
print("RBF Kernel Confusion Matrix:\n", confusion_matrix(y_test, y_pred_rbf))
print("RBF Kernel Classification Report:\n", classification_report(y_test, y_pred_rbf))
```

```python
poly = SVC(kernel='poly')
poly.fit(X_train, y_train)
y_pred_poly = poly.predict(X_test)
print("RBF Kernel Accuracy:", accuracy_score(y_test, y_pred_poly))
print("RBF Kernel Confusion Matrix:\n", confusion_matrix(y_test, y_pred_poly))
print("RBF Kernel Classification Report:\n", classification_report(y_test, y_pred_poly))
```
