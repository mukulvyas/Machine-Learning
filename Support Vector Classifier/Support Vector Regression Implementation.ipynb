```python
## here those are categorical but in binary form we'll do feature encoding
# two types Label Encoding and One Hot Encoding
# Label Encoding is used for ordinal data and One Hot Encoding is used for nominal data 
# if we have more than two categories then we use One Hot Encoding and if we have two categories then we can use Label Encoding

## independent and dependent variables
X = df[['tip', 'sex', 'smoker', 'day', 'time', 'size']]
y = df['total_bill']
```

```python
## train set and test set
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=10) 
```

```python
X_train.head()
```

```python
## now we have to do feature encoding
## we should divide or do the train test split before feature encoding so that we can avoid data leakage, 
## model should not know about the test data during training

from sklearn.preprocessing import LabelEncoder
```

```python
le1 = LabelEncoder()
le2 = LabelEncoder()
le3 = LabelEncoder()
```

```python
X_train['sex']=le1.fit_transform(X_train['sex'])
X_train['smoker']=le2.fit_transform(X_train['smoker'])
X_train['smoker']=le3.fit_transform(X_train['smoker'])
```

```python
X_train.head()
```

```python
X_test['sex']=le1.transform(X_test['sex'])
X_test['smoker']=le2.transform(X_test['smoker'])
X_test['smoker']=le3.transform(X_test['smoker'])
```

```python
X_test.head()
```

```python
## One Hot Encoding
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
## we give tuple of column names and then we give the transformer
ct = ColumnTransformer(transformers=[('onehot',OneHotEncoder(drop='first'),[3])
                                     ],remainder='passthrough')
```

```python
ct = ColumnTransformer(
    transformers=[
        ('onehot', OneHotEncoder(drop='first'), [3, 4])
    ],
    remainder='passthrough'
)
X_train = ct.fit_transform(X_train)
```

```python
X_train
```

```python
X_test = ct.transform(X_test)
```

```python
## SVR  Support Vector Regression
from sklearn.svm import SVR
svr = SVR()
```

```python
svr.fit(X_train, y_train)
```

```python
y_pred = svr.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = mse ** 0.5
r2 = r2_score(y_test, y_pred)
print(f"MAE: {mae}")
print(f"MSE: {mse}")
print(f"RMSE: {rmse}")
print(f"R2: {r2}")
```

```python
## Hyperparameter Tuning
from sklearn.model_selection import GridSearchCV
param_grid = {
    'kernel': [ 'rbf'],
    'C': [0.1, 1, 10,100,1000],
    'gamma': [0.01, 0.0001, 0.001, 0.1, 1],
}
```

```python
grid = GridSearchCV(SVR(), param_grid, refit = True,verbose = 2)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid_prediction = grid.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
mae = mean_absolute_error(y_test, grid_prediction)
mse = mean_squared_error(y_test, grid_prediction)
rmse = mse ** 0.5
r2 = r2_score(y_test, grid_prediction)
print(f"MAE: {mae}")
print(f"MSE: {mse}")
print(f"RMSE: {rmse}")
print(f"R2: {r2}")
```
