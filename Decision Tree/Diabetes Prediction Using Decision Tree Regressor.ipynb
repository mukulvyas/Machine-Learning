## Diabetes Predictions

```python
from sklearn.datasets import load_diabetes
```

```python
dataset = load_diabetes()
```

```python
dataset
```

```python
print(dataset['DESCR'])
```

```python
import pandas as pd
df_diabetes = pd.DataFrame(data=dataset.data,columns = dataset.feature_names)
```

```python
# indepedent and dependent features
X = df_diabetes
y = dataset.target
```

```python
X
```

```python
y
```

```python
## train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3, random_state=10)
```

```python
X_train.head()
```

```python
## checking the correlation
X_train.corr()
```

```python
## let check the heatmap
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(15,10))
sns.heatmap(X_train.corr(),annot=True)

```

```python
## let's train the model

from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor()
regressor.fit(X_train,y_train)
```

### Hyperparameter Tunning

```python
param ={
    'criterion' :['squared_error','friedman_mse','absolute_error','poisson'],
    'splitter': ['best', 'random'],
    'max_depth': [1,2,3,4,5,10,15,20,25],
    'max_features': ['auto', 'sqrt', 'log2'],
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
regressor = DecisionTreeRegressor()
```

```python
grid = GridSearchCV(estimator = regressor,param_grid = param, scoring ='neg_mean_squared_error')
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred = grid.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
print("R2 Score:", r2_score(y_test, y_pred))
print("Mean Absolute Error:", mean_absolute_error(y_test, y_pred))
```

```python
selectmodel = DecisionTreeRegressor(criterion= 'squared_error',
 max_depth= 4,
 max_features= 'log2',
 splitter= 'best')
```

```python
selectmodel.fit(X_train,y_train)
```

```python
## Visualizing the Decision Tree
from sklearn import tree
plt.figure(figsize=(20,10))
tree.plot_tree(selectmodel, filled=True, feature_names=X_train.columns, fontsize=10)
```
